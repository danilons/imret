```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from imret.query import KnowledgeBase
from imret.dataset import Dataset
```

```python
dataset = Dataset('../data/datasets/', 'test', '../data/images')
```

```python
df = pd.read_csv('../data/preposition/index.csv')
df.head()
```

```python
kb = KnowledgeBase(df, ltb_runner='../data/prover/E/PROVER/e_ltb_runner',
                   eprover='../data/prover/E/PROVER/eprover',
                   sumo='/home/danilo/sigma_run/KBs/SUMO.tptp')
```

```python
query = 'car-on-road'
```

```python
image = list(set(df[(df['object1'] == 'car') & (df['object2'] == 'road') & (df['preposition'] == 'on')].image))[0]
imarray = dataset.get_im_array(image, rgb=True)
plt.imshow(imarray)
```

```python
_, is_proved, log = kb.prover(image=image, query=query)
print "Proof: ", is_proved
print log
```
